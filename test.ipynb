```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    'Sample': ['P4_rep1', 'P4_rep2', 'P6_rep1', 'P6_rep2'],
    'MQuad': [3, 3, 3, 4],
    'betaMQuad': [5, 4, 2, 6]
}

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to long format for seaborn
df_melted = df.melt(id_vars='Sample', value_vars=['MQuad', 'betaMQuad'],
                    var_name='Method', value_name='Variants')

# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Method', y='Variants', data=df_melted, palette='Set2')

# Add individual data points
sns.stripplot(x='Method', y='Variants', data=df_melted, color='black', alpha=0.7)

# Customize the plot
plt.title('Distribution of Variants Detected by MQuad and betaMQuad')
plt.xlabel('Method')
plt.ylabel('Number of Variants')

# Show the plot
plt.tight_layout()
plt.show()
```
